# RGB

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

def hitung_histogram_gambar(gambar):
    # Menghitung histogram untuk setiap saluran warna (R, G, B)
    hist_r = cv2.calcHist([gambar], [0], None, [256], [0, 256])
    hist_g = cv2.calcHist([gambar], [1], None, [256], [0, 256])
    hist_b = cv2.calcHist([gambar], [2], None, [256], [0, 256])

    return hist_r, hist_g, hist_b

def hitung_histogram_folder(direktori):
    # Inisialisasi array untuk menyimpan histogram total
    hist_total_r = np.zeros((256, 1))
    hist_total_g = np.zeros((256, 1))
    hist_total_b = np.zeros((256, 1))

    # Dapatkan daftar file dalam direktori
    file_list = os.listdir(direktori)

    # Melakukan perhitungan histogram untuk setiap gambar
    for file_name in file_list:
        # Baca gambar
        gambar = cv2.imread(os.path.join(direktori, file_name))

        # Hitung histogram gambar
        hist_r, hist_g, hist_b = hitung_histogram_gambar(gambar)

        # Tambahkan histogram gambar ke histogram total
        hist_total_r += hist_r
        hist_total_g += hist_g
        hist_total_b += hist_b

    return hist_total_r, hist_total_g, hist_total_b

def visualisasikan_histogram(hist_r, hist_g, hist_b):
    # Menampilkan histogram menggunakan matplotlib
    plt.figure(figsize=(9, 2))
    plt.subplot(1, 3, 1)
    plt.plot(hist_r, color='red')
    plt.title('R')
    plt.xlabel('Pixel')
    plt.ylabel('Frekeunsi')

    plt.subplot(1, 3, 2)
    plt.plot(hist_g, color='green')
    plt.title('G')
    plt.xlabel('Pixel')
    plt.ylabel('Frekuensi')

    plt.subplot(1, 3, 3)
    plt.plot(hist_b, color='blue')
    plt.title('B')
    plt.xlabel('Pixel')
    plt.ylabel('Frekuensi')

    plt.tight_layout()
    plt.show()
```

```python
# Lets see
print("Sebelum segmentasi")
direktori = "T:/PCD/Banana_Condensed_Three_Cat/underripe"
hist_r, hist_g, hist_b = hitung_histogram_folder(direktori)
visualisasikan_histogram(hist_r, hist_g, hist_b)

print("\nSesudah segmentasi")
direktori = "T:/PCD/Banana_Condensed_Three_Cat/segurp"
hist_r, hist_g, hist_b = hitung_histogram_folder(direktori)
visualisasikan_histogram(hist_r, hist_g, hist_b)
```

# HSV

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

def hitung_histogram_gambar(gambar):
    # Konversi gambar ke ruang warna HSV
    hsv = cv2.cvtColor(gambar, cv2.COLOR_BGR2HSV)

    # Menghitung histogram untuk setiap saluran warna (H, S, V)
    hist_h = cv2.calcHist([hsv], [0], None, [180], [0, 180])
    hist_s = cv2.calcHist([hsv], [1], None, [256], [0, 256])
    hist_v = cv2.calcHist([hsv], [2], None, [256], [0, 256])

    return hist_h, hist_s, hist_v

def hitung_histogram_folder(direktori):
    # Inisialisasi array untuk menyimpan histogram total
    hist_total_h = np.zeros((180, 1))
    hist_total_s = np.zeros((256, 1))
    hist_total_v = np.zeros((256, 1))

    # Dapatkan daftar file dalam direktori
    file_list = os.listdir(direktori)

    # Melakukan perhitungan histogram untuk setiap gambar
    for file_name in file_list:
        # Baca gambar
        gambar = cv2.imread(os.path.join(direktori, file_name))

        # Hitung histogram gambar
        hist_h, hist_s, hist_v = hitung_histogram_gambar(gambar)

        # Tambahkan histogram gambar ke histogram total
        hist_total_h += hist_h
        hist_total_s += hist_s
        hist_total_v += hist_v

    return hist_total_h, hist_total_s, hist_total_v

def visualisasikan_histogram(hist_h, hist_s, hist_v):
    # Menampilkan histogram menggunakan matplotlib
    plt.figure(figsize=(9, 2))
    plt.subplot(1, 3, 1)
    plt.plot(hist_h, color='red')
    plt.title('Hue')
    plt.xlabel('Pixel')
    plt.ylabel('Frekuensi')

    plt.subplot(1, 3, 2)
    plt.plot(hist_s, color='green')
    plt.title('Saturation')
    plt.xlabel('Pixel')
    plt.ylabel('Frekuensi')

    plt.subplot(1, 3, 3)
    plt.plot(hist_v, color='blue')
    plt.title('Value')
    plt.xlabel('Pixel')
    plt.ylabel('Frekuensi')

    plt.tight_layout()
    plt.show()
```

# LAB

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

def hitung_histogram_gambar(gambar):
    # Konversi gambar ke ruang warna LAB
    lab = cv2.cvtColor(gambar, cv2.COLOR_BGR2LAB)

    # Menghitung histogram untuk setiap saluran warna (L, A, B)
    hist_l = cv2.calcHist([lab], [0], None, [256], [0, 256])
    hist_a = cv2.calcHist([lab], [1], None, [256], [0, 256])
    hist_be = cv2.calcHist([lab], [2], None, [256], [0, 256])

    return hist_l, hist_a, hist_be

def hitung_histogram_folder(direktori):
    # Inisialisasi array untuk menyimpan histogram total
    hist_total_l = np.zeros((256, 1))
    hist_total_a = np.zeros((256, 1))
    hist_total_be = np.zeros((256, 1))

    # Dapatkan daftar file dalam direktori
    file_list = os.listdir(direktori)

    # Melakukan perhitungan histogram untuk setiap gambar
    for file_name in file_list:
        # Baca gambar
        gambar = cv2.imread(os.path.join(direktori, file_name))

        # Hitung histogram gambar
        hist_l, hist_a, hist_be = hitung_histogram_gambar(gambar)

        # Tambahkan histogram gambar ke histogram total
        hist_total_l += hist_l
        hist_total_a += hist_a
        hist_total_be += hist_be

    return hist_total_l, hist_total_a, hist_total_be

def visualisasikan_histogram(hist_l, hist_a, hist_be):
    # Menampilkan histogram menggunakan matplotlib
    plt.figure(figsize=(9, 2))
    plt.subplot(1, 3, 1)
    plt.plot(hist_l, color='black')
    plt.title('L')
    plt.xlabel('Pixel')
    plt.ylabel('Frekuensi')

    plt.subplot(1, 3, 2)
    plt.plot(hist_a, color='green')
    plt.title('A')
    plt.xlabel('Pixel')
    plt.ylabel('Frekuensi')

    plt.subplot(1, 3, 3)
    plt.plot(hist_b, color='blue')
    plt.title('B*')
    plt.xlabel('Pixel')
    plt.ylabel('Frekuensi')

    plt.tight_layout()
    plt.show()
```
